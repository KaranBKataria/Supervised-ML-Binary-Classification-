## Feature Engineering:

```python
df.diagnosis = np.select([df['diagnosis'] == 'M', df['diagnosis'] == 'B'], [1,0], default = 2)
df
```

```python
y = df.diagnosis
X = df[['radius_mean', 'perimeter_mean', 'area_mean', 'concavity_mean', 'concave points_mean', 'perimeter_se', 'area_se', 'radius_worst', 'perimeter_worst', 'area_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst']]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
print(X.mean(axis=0))
```

```python
print(X.std(axis=0))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state= 27)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predicted)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predicted)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, y_predicted)
```
